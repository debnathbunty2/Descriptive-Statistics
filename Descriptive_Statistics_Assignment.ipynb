5. Dispersion Introduction (Medium)
A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40].
What does the range tell us about the variation in the student’s puzzle-solving time?

- The range is a measure of dispersion in statistics that quantifies the spread of a data set by calculating the difference between the maximum and minimum values. In this case, the data set is.

The maximum value (longest time) is 40.
The minimum value (shortest time) is 25.
The range is calculated as Maximum -Minimum =40-25=15

A larger range would indicate greater variation in the student's times, while a smaller range would suggest more consistent times. The range of 15 minutes indicates that the total spread of the student's puzzle-solving times is 15 minutes.

7. Q7: Variance for Decision-Making (Medium)
Two delivery companies track delivery delays (in minutes).

Company A: variance = 6

Company B: variance = 15
 Which company is more consistent, and why?

- Company A is more consistent.
Explanation:
Consistency in data refers to how close the data points are to the mean or average value. Variance is a measure of the spread or dispersion of data points from the mean. A lower variance indicates that the data points are clustered more closely around the mean, implying greater consistency.
In this case:
Company A has a variance of 6.
Company B has a variance of 15.
Since Company A has a lower variance (6 is less than 15), its delivery delays are less spread out and closer to the average delay time compared to Company B. Therefore, Company A provides a more consistent delivery service.

 - Coin B is riskier to invest in.

Therefore, Coin B, with a standard deviation of $120, has a wider range of potential price changes than Coin A, which has a standard deviation of $30. This makes Coin B the riskier investment.
